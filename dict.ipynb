```python
# Initialize a dictionary
person = {
    "name": "John",
    "age": 30,
    "gender": "Male",
    "city": "New York"
}

print("Original Dictionary:", person)

# 1. Access a value using a key
Name=person['name']
print(Name)

# 2. Add a new key-value pair
person["occupation"] = "Engineer"
print("After adding occupation:", person)

# 3. Update an existing value
person["age"] = 31
print("After updating age:", person)

# 4. Remove a key-value pair using pop
removed_value = person.pop("city")
print("After removing city:", person, "| Removed Value:", removed_value)

# 5. Remove the last inserted key-value pair using popitem
removed_item = person.popitem()
print("After popitem:", person, "| Removed Item:", removed_item)

# 6. Check if a key exists
key = "gender"
print(f"Is '{key}' present in the dictionary?", key in person)

# 7. Get the keys of the dictionary
print("Keys:", person.keys())

# 8. Get the values of the dictionary
print("Values:", person.values())

# 9. Get all key-value pairs as tuples
print("Items:", person.items())

# 10. Copy the dictionary
person_copy = person.copy()
print("Copied Dictionary:", person_copy)

# 11. Clear all items from the dictionary
person_copy.clear()
print("After clearing the copied dictionary:", person_copy)

# 12. Set default value for a key
default_gender = person.setdefault("gender", "Unknown")
print("Gender:", default_gender, "| Updated Dictionary:", person)

# 13. Merge two dictionaries
additional_info = {"hobby": "Photography", "pets": "None"}
person.update(additional_info)
print("After merging with additional_info:", person)

# 14. Create a dictionary from keys
keys = ["a", "b", "c"]
default_value = 0
new_dict = dict.fromkeys(keys, default_value)
print("New dictionary from keys:", new_dict)

# 15. Iterate over the dictionary
print("Iterating over dictionary:")
for key, value in person.items():
    print(f"{key}: {value}")
```
