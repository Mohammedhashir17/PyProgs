```python
import pandas as pd
import matplotlib.pyplot as plt

data={'Patient':['Alice','Bob','John','Mia','Dom'],'Age':[21,15,40,35,60],'Weight':[60,80,72,88,40],'Systolic BP':[80,110,120,100,90],'Diastolic BP':[90,100,120,110,80]}
df=pd.DataFrame(data)

print("Original Dataframe")
print(df)

df['Systolic BP']=pd.to_numeric(df['Systolic BP'])
df['Diastolic BP']=pd.to_numeric(df['Diastolic BP'])

df['Height']=1.75
df['BMI']=df['Weight']/(df['Height']*df['Height'])

print("Modified Dataframe")
print(df)

df.plot(kind='bar',x='Patient',y='BMI',title='Bar Chart')
plt.xlabel("Patient")
plt.ylabel("BMI")
plt.show()

df.plot.pie(y='Systolic BP',labels=df['Patient'],autopct='%1.1f%%',title="Pie chart")
plt.show()

df.plot(kind='scatter',x='BMI',y='Systolic BP',title='Scatter Plot')
plt.xlabel("BMI")
plt.ylabel("Systolic BP")
plt.show()
```
